```python
# 1. Convert to Datetime
# Using format='%d-%m-%Y' to  DD-MM-YYYY format
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%d-%m-%Y')
data['Ship Date'] = pd.to_datetime(data['Ship Date'], format='%d-%m-%Y')

print("--- Datetime Conversion Complete ---")
print(data[['Order Date', 'Ship Date']].dtypes)
print("\n")


# 2. Quality Checks

# Checking missing values
print("missing values ")
missing_values = data.isnull().sum()
print(missing_values)
#  if find missing value can use df.dropna() to remove those rows 
# or df.fillna() to impute them.
print("\n")

# Check for and handle duplicates
print(" Duplicate Rows ")
duplicate_count = data.duplicated().sum()
print(f"Total duplicate rows found: {duplicate_count}")

if duplicate_count > 0:
    # This will automatically keep first occurrence and drop the rest
    data = data.drop_duplicates()
    print("Duplicate rows have been successfully dropped.")
print("\n")

# Verify data types (float64 or int64)
print("Verify Data Types")
print(data.dtypes)
```

```python
# Step 3: Extract Time Features
print("Extracting time features")

# Extract standard date components using .dt 
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month
data['Week'] = data['Order Date'].dt.isocalendar().week
data['DayOfWeek'] = data['Order Date'].dt.dayofweek  # Monday=0, Sunday=6
data['Quarter'] = data['Order Date'].dt.quarter

# Create custom function to determine the Season based on the month
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# function to create a new 'Season' column
data['Season'] = data['Month'].apply(get_season)

print("time features successfully extracted")
# Display the first 5 rows of our new columns to verify
display(data[['Order Date', 'Year', 'Month', 'Week', 'DayOfWeek', 'Quarter', 'Season']].head(10))
```

```python
# Step 4: Time-Series Aggregation
print("--- Aggregating Sales Data ---")

# use pandas resampling, we must set the Datetime column as the index
# We create new dataframe specifically for time-series task
ts_data = data.set_index('Order Date').sort_index()

# Aggregate daily sales into monthly and weekly total
# 'ME' stands for month-rnd frequency, 'W' stands for Weekly frequency
monthly_sales = ts_data['Sales'].resample('ME').sum()
weekly_sales = ts_data['Sales'].resample('W').sum()

print("monthly Sales (First 10 rows):")
print(monthly_sales.head(10))
print("\nWeekly ssales (First 10 rows):")
print(weekly_sales.head(10))
```

```python
# Task 2: Time Series Analysis & Decomposition

from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller

print("-- 1. Overall Monthly Sales Trend --")
# plotting the overall monthly sales trend across for all  4 years
plt.figure(figsize=(12, 5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', color='teal', linewidth=2)
plt.title('Overall Monthly Sales Trend (4 Years)')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig("chart/overall_monthly_sales_trend(4_years).png")
plt.show()

print("\n--- 2. Time Series Decomposition ---")
# Apply Time Series Decomposition (breaking the signal into trend, seasonal, and residual)
# We use a 'additive' model since the magnitude of seasonality doesn't change much over time
decomposition = seasonal_decompose(monthly_sales, model='additive')

# Plot all 4 components clearly on one figure
fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.savefig("chart/time_series_destribution.png")
plt.show()

print("\n--3. Stationarity Check: Augmented Dickey-Fuller (ADF) Test--")
# Define a reusable function for the ADF test
def run_adf_test(series, title=""):
    print(f"ADF Test Results {title}")
    result = adfuller(series.dropna())
    print(f'ADF Statistic: {result[0]:.4f}')
    print(f'p-value: {result[1]:.4f}')
    print('critical values:')
    for key, value in result[4].items():
        print(f'   {key}: {value:.4f}')
        
    # Standard threshold is 0.05
    if result[1] <= 0.05:
        print("Conclusion: The series is STATIONARY (Reject the null hypothesis).\n")
    else:
        print("Conclusion: The series is NON-STATIONARY (Fail to reject the null hypothesis).\n")

# runing  test on our original monthly sales data
run_adf_test(monthly_sales, title="(Original Data)")

print("-- 4. Applying Differencing --")
# Time-series forecasting models often require stationary data. 
# If the p-value above is > 0.05, we apply first-order differencing and test again.
monthly_sales_diff = monthly_sales.diff().dropna()

# Plot the differenced data to see the effect visually
plt.figure(figsize=(12, 4))
plt.plot(monthly_sales_diff.index, monthly_sales_diff.values, color='purple', linewidth=1.5)
plt.title('First-Order Differenced Monthly Sales')
plt.axhline(0, color='black', linestyle='--')
plt.savefig("first_order_differeced_monthly_sales,png")
plt.show()

# re-test  differenced data
run_adf_test(monthly_sales_diff, title="(First-Order differenced Data)")
```

2. Stationarity & The ADF Test
To build statistical forecasting models like SARIMA in the next task, the data generally needs to be "stationary" (meaning its average and variance don't wildly shift over time).

I ran the Augmented Dickey-Fuller (ADF) test, and the results were actually pretty surprising:

Usually, retail sales data has such a strong upward trend that it fails this test until we difference it. However, the test on our Original Data returned an ADF Statistic of -5.3008 and a p-value of 0.0000.

Because the p-value is well below the standard 0.05 cutoff, we can reject the null hypothesis immediately.

<b>Conclusion</b>: Our baseline monthly sales data is already stationary. Even though I also applied first-order differencing just to check the workflow (which also passed perfectly), the original data is stable enough that we can move forward into the forecasting phase without needing aggressive transformations.

```python
# Model libraries
#importing all nessesary all liberary that needed to perfom all model task 
from statsmodels.tsa.statespace.sarimax import SARIMAX
from prophet import Prophet
import xgboost as xgb

# evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings('ignore') # keep notebook clean from statsmodels convergence warnings

# Define a custom MAPE function since sklearn doesn't have a robust one for this
def calculate_mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

print("--Preparing Data for Model Evaluation--")
# Split data: Train on everything except the last 3 months of 2018
train_series = monthly_sales.iloc[:-3]
test_series = monthly_sales.iloc[-3:]
print(f"training months: {len(train_series)} | Testing months: {len(test_series)}")
```

Model 1: SARIMA
This block handles the traditional statistical approach. Since we already proved the data is stationary in Task 2, we can set d=0

```python
# ---------------------------------------------------------
# MODEL 1: SARIMA
# ---------------------------------------------------------
print("--Training SARIMA Model --")

# using (1,0,1) because data is stationary, and seasonal (1,1,1,12) for yearly cycle
sarima_model = SARIMAX(train_series, 
                       order=(1, 0, 1), 
                       seasonal_order=(1, 1, 1, 12), 
                       enforce_stationarity=False)
sarima_fit = sarima_model.fit(disp=False)

# 1. evaluate on 3-month test set
sarima_pred = sarima_fit.predict(start=len(train_series), end=len(train_series)+2)
sarima_mae = mean_absolute_error(test_series, sarima_pred)
sarima_rmse = np.sqrt(mean_squared_error(test_series, sarima_pred))
sarima_mape = calculate_mape(test_series, sarima_pred)

print(f"SARIMA Test Metrics -> MAE: {sarima_mae:.2f} | RMSE: {sarima_rmse:.2f} | MAPE: {sarima_mape:.2f}%")

# 2. Fit on FULL dataset to predict actual unknown future (Q1 2019)
sarima_full_fit = SARIMAX(monthly_sales, order=(1, 0, 1), seasonal_order=(1, 1, 1, 12), enforce_stationarity=False).fit(disp=False)
sarima_future = sarima_full_fit.predict(start=len(monthly_sales), end=len(monthly_sales)+2)

# Plot the forecast
plt.figure(figsize=(10, 4))
plt.plot(monthly_sales.index, monthly_sales.values, label='Actual Sales (2015-2018)')
plt.plot(sarima_future.index, sarima_future.values, color='red', marker='o', linewidth=2, label='SARIMA 3-Month Forecast')
plt.title('SARIMA Model - future Forecast (Q1 2019)')
plt.ylabel('Total Sales ($)')
plt.legend()
plt.savefig("chart/SARIMA_Model_Future_Forecast_(Q1_2019).png")
plt.show()
```

Model 2: Facebook Prophet
Prophet is strict about its data format, so we create a specialized dataframe just

```python
# ---------------------------------------------------------
# MODEL 2: Facebook Prophet
# ---------------------------------------------------------
print("--Training Prophet Model--")

# prophet strictly requires columns to be named as 'ds' (datestamp) and 'y' (target)
prophet_df = monthly_sales.reset_index()
prophet_df.columns = ['ds', 'y']

prophet_train = prophet_df.iloc[:-3]
prophet_test = prophet_df.iloc[-3:]

# Initializ model explicitly telling it to look for yearly repeating patterns
m = Prophet(yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False)
m.fit(prophet_train)

# 1. Evaluate on the test set
future_eval = m.make_future_dataframe(periods=3, freq='ME') 
forecast_eval = m.predict(future_eval)
prophet_pred = forecast_eval['yhat'].iloc[-3:].values

prophet_mae = mean_absolute_error(test_series, prophet_pred)
prophet_rmse = np.sqrt(mean_squared_error(test_series, prophet_pred))
prophet_mape = calculate_mape(test_series.values, prophet_pred)

print(f"Prophet Test Metrics -> MAE: {prophet_mae:.2f} | RMSE: {prophet_rmse:.2f} | MAPE: {prophet_mape:.2f}%")

# 2. Fit on FULL dataset for the true future forecast
m_full = Prophet(yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False)
m_full.fit(prophet_df)
future_3mo = m_full.make_future_dataframe(periods=3, freq='ME')
forecast_future = m_full.predict(future_3mo)

# Plot Prophet's forecast and its built-in trend/seasonality breakdown
fig1 = m_full.plot(forecast_future)
plt.title('Prophet Model - Future Forecast (Q1 2019)')
plt.savefig("Prophet Model - future Forecast (Q1 2019).png")
plt.show()

print("breaking down prophet's trend and Seasonality Components:")
fig2 = m_full.plot_components(forecast_future)
plt.show()
```

Model 3: XGBoost (Supervised ML Approach) : To use a tree-based ensemble on time-series data, we have to manually build the "memory" into the rows by creating lag features.

```python
# ---------------------------------------------------------
# MODEL 3: XGBoost
# ---------------------------------------------------------
print("--- Training XGBoost Model ---")

# We need to transform the time series into standard ML dataframe form.
xgb_df = monthly_sales.to_frame(name='Sales')

# Creat lag features (looking back 1, 2, and 3 months)
xgb_df['Lag_1'] = xgb_df['Sales'].shift(1)
xgb_df['Lag_2'] = xgb_df['Sales'].shift(2)
xgb_df['Lag_3'] = xgb_df['Sales'].shift(3)

# Create a 3-month rolling mean to capture short-term momentum
xgb_df['Rolling_Mean_3'] = xgb_df['Sales'].shift(1).rolling(window=3).mean()

# Add time features
xgb_df['Month'] = xgb_df.index.month
xgb_df['Quarter'] = xgb_df.index.quarter

# Drop the first 3 rows so they will have NaN values from shifting
xgb_df = xgb_df.dropna()

# Define features (X) and target (y)
features = ['Lag_1', 'Lag_2', 'Lag_3', 'Rolling_Mean_3', 'Month', 'Quarter']
X = xgb_df[features]
y = xgb_df['Sales']

# 1. Train and Test and  split for evaluation
X_train, X_test = X.iloc[:-3], X.iloc[-3:]
y_train, y_test = y.iloc[:-3], y.iloc[-3:]

xgb_model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
xgb_model.fit(X_train, y_train)

# predict and evaluate
xgb_pred = xgb_model.predict(X_test)
xgb_mae = mean_absolute_error(y_test, xgb_pred)
xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_pred))
xgb_mape = calculate_mape(y_test.values, xgb_pred)

print(f"XGBoost Test metrics -> MAE: {xgb_mae:.2f} | RMSE: {xgb_rmse:.2f} | MAPE: {xgb_mape:.2f}%")

# 2. Forecast actual future (iterative prediction since we need future lags)
# Train on full dataset
xgb_full = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
xgb_full.fit(X, y)

# We have to build the future dataframe row by row
last_known_data = xgb_df.iloc[-1:].copy()
future_dates = pd.date_range(start=xgb_df.index[-1] + pd.DateOffset(months=1), periods=3, freq='ME')
future_predictions = []

current_lag1 = last_known_data['Sales'].values[0]
current_lag2 = last_known_data['Lag_1'].values[0]
current_lag3 = last_known_data['Lag_2'].values[0]
# calculate rolling mean based last 3 actual sales months
sales_history = list(y.iloc[-3:].values) 

for date in future_dates:
    current_rolling = np.mean(sales_history[-3:])
    
    # Create the feature row for the new month
    next_month_features = pd.DataFrame({
        'Lag_1': [current_lag1],
        'Lag_2': [current_lag2],
        'Lag_3': [current_lag3],
        'Rolling_Mean_3': [current_rolling],
        'Month': [date.month],
        'Quarter': [date.quarter]
    })
    
    # Predict the month
    pred = xgb_full.predict(next_month_features)[0]
    future_predictions.append(pred)
    
    # Update lags and history for the next loop iteration
    sales_history.append(pred)
    current_lag3 = current_lag2
    current_lag2 = current_lag1
    current_lag1 = pred

# Plot the XGBoost Forecast
plt.figure(figsize=(10, 4))
plt.plot(monthly_sales.index, monthly_sales.values, label='Actual Sales (2015-2018)')
plt.plot(future_dates, future_predictions, color='green', marker='s', linewidth=2, label='XGBoost 3-Month Forecast')
plt.title('XGBoost Model - Future Forecast (Q1 2019)')
plt.ylabel('Total Sales ($)')
plt.legend()
plt.savefig("chart/XGBoost_Model_Future_Forecast_(Q1_2019).png")
plt.show()
```

```python
# Task 4: Product Category & Region Level Forecasting (Bulletproof Version)
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings
import os
warnings.filterwarnings('ignore')

print("--- Safely Reloading Data ---")
# 1. Force a fresh load of the data to guarantee it is intact
fresh_df = pd.read_csv('archive/train.csv')

# 2. Re-apply the datetime conversion so the resampling works
fresh_df['Order Date'] = pd.to_datetime(fresh_df['Order Date'], format='%d-%m-%Y')

print("--- Running SARIMA Forecasts for Business Segments ---")
# 3. Define segments using this guarante fresh data
segments = {
    'Furniture Category': fresh_df[fresh_df['Category'] == 'Furniture'],
    'Technology Category': fresh_df[fresh_df['Category'] == 'Technology'],
    'Office Supplies Category': fresh_df[fresh_df['Category'] == 'Office Supplies'],
    'West Region': fresh_df[fresh_df['Region'] == 'West'],
    'East Region': fresh_df[fresh_df['Region'] == 'East']
}

future_forecasts = {}

# 4. loop through each segment and forecast
for name, segment_df in segments.items():
    print(f"Training model for: {name}")
    
    # Aggregate to monthly time-series
    ts_data = segment_df.set_index('Order Date').sort_index()
    monthly_ts = ts_data['Sales'].resample('ME').sum()
    
    # Fit the winning SARIMA model: (1,0,1) x (1,1,1,12)
    model = SARIMAX(monthly_ts, order=(1, 0, 1), seasonal_order=(1, 1, 1, 12), enforce_stationarity=False)
    fit_model = model.fit(disp=False)
    
    # predict the next 3 months (Jan, Feb, Mar 2019)
    forecast = fit_model.predict(start=len(monthly_ts), end=len(monthly_ts)+2)
    future_forecasts[name] = forecast

print("\n--- Plotting Segment Comparison ---")

# Plot all 5 forecasts together on one comparison chart
plt.figure(figsize=(12, 6))

markers = ['o', 's', '^', 'D', 'x']
for i, (name, forecast) in enumerate(future_forecasts.items()):
    plt.plot(forecast.index, forecast.values, marker=markers[i], linewidth=2.5, label=name)

plt.title('SARIMA 3-Month Forecast Comparison by Segment (Q1 2019)', fontsize=14)
plt.xlabel('Forecast Horizon', fontsize=12)
plt.ylabel('Projected Sales ($)', fontsize=12)
plt.xticks(future_forecasts['Furniture Category'].index, ['Month 1 (Jan)', 'Month 2 (Feb)', 'Month 3 (Mar)'])
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Save and show the chart
os.makedirs("charts", exist_ok=True)
plt.savefig("charts/SARIMA_3-Month_Forecast_Comparison_by_Segment_(Q1_2019).png")
plt.show()

# Display the final forecasted values for accurate reporting
forecast_df = pd.DataFrame(future_forecasts)
forecast_df.index = ['Month 1 (Jan)', 'Month 2 (Feb)', 'Month 3 (Mar)']
print("\nProjected Sales Figures for Q1 2019:")
display(forecast_df.round(2))
```

```python
# Task 5: Anomaly Detection in Sales Data (Bulletproof Version)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import warnings
import os
warnings.filterwarnings('ignore')

print("--- Safely Reloading Data for Task 5 ---")
# 1. Force a fresh load of the data to guarantee it is intact
fresh_df = pd.read_csv('archive/train.csv')
fresh_df['Order Date'] = pd.to_datetime(fresh_df['Order Date'], format='%d-%m-%Y')

print("--- Preparing Weekly Sales Data ---")
# Safely create a fresh weekly sales dataframe
ts_data = fresh_df.set_index('Order Date').sort_index()
weekly_sales = ts_data['Sales'].resample('W').sum().to_frame()

# ---------------------------------------------------------
# Method 1: Isolation Forest (Machine Learning)
# ---------------------------------------------------------
print("Running Isolation Forest...")
# Contamination defines the proportion of outliers we expect (setting it to 5% of our weeks)
iso_forest = IsolationForest(contamination=0.05, random_state=42)
# The model returns -1 for anomalies, and 1 for normal data points
weekly_sales['Iso_Anomaly'] = iso_forest.fit_predict(weekly_sales[['Sales']])

# ---------------------------------------------------------
# Method 2: Z-Score (Statistical)
# ---------------------------------------------------------
print("Running Z-Score Detection...")
# Calculate a 4-week rolling mean and standard deviation
weekly_sales['Rolling_Mean'] = weekly_sales['Sales'].rolling(window=4).mean()
weekly_sales['Rolling_Std'] = weekly_sales['Sales'].rolling(window=4).std()

# Calculate the Z-Score (how many standard deviations away from the rolling mean)
weekly_sales['Z_Score'] = (weekly_sales['Sales'] - weekly_sales['Rolling_Mean']) / weekly_sales['Rolling_Std']

# Flag anomalies where the Z-Score is > 2 (massive spike) or < -2 (massive drop)
# We fill NaN values (from the first few rolling weeks) with 0 so they aren't flagged
weekly_sales['Z_Anomaly'] = np.where((weekly_sales['Z_Score'] > 2) | (weekly_sales['Z_Score'] < -2), -1, 1)

# ---------------------------------------------------------
# Plotting and Comparison
# ---------------------------------------------------------
print("\n--- Plotting Anomalies ---")
plt.figure(figsize=(14, 6))

# Plot the baseline weekly sales
plt.plot(weekly_sales.index, weekly_sales['Sales'], label='Weekly Sales', color='teal', alpha=0.6, linewidth=1.5)

# Extract and plot the anomalies for both methods
iso_outliers = weekly_sales[weekly_sales['Iso_Anomaly'] == -1]
z_outliers = weekly_sales[weekly_sales['Z_Anomaly'] == -1]

# Plot Isolation Forest Anomalies (Red 'X')
plt.scatter(iso_outliers.index, iso_outliers['Sales'], color='red', s=80, marker='x', label='Isolation Forest Anomaly', zorder=5)

# Plot Z-Score Anomalies (Orange Circles)
plt.scatter(z_outliers.index, z_outliers['Sales'], color='orange', s=80, marker='o', facecolors='none', linewidths=2, label='Z-Score Anomaly (Rolling > 2 Std Dev)', zorder=5)

plt.title('Sales Anomaly Detection: Isolation Forest vs. Z-Score', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Weekly Sales ($)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

# Save the chart for your final ZIP submission
os.makedirs("charts", exist_ok=True)
plt.savefig("charts/Anomaly_Detection_Comparison.png")
plt.show()

# Display the exact dates that were flagged to figure out WHY they happened
print("\n--- Detected Anomalies Analysis ---")
print(f"Total Isolation Forest Anomalies Detected: {len(iso_outliers)}")
print(f"Total Z-Score Anomalies Detected: {len(z_outliers)}")

# Combine them to see where they overlap or disagree
comparison_df = weekly_sales[(weekly_sales['Iso_Anomaly'] == -1) | (weekly_sales['Z_Anomaly'] == -1)][['Sales', 'Iso_Anomaly', 'Z_Anomaly', 'Z_Score']]
comparison_df = comparison_df.replace({-1: 'FLAGGED', 1: 'Normal'})
print("\nDetailed Breakdown of Flagged Weeks:")
display(comparison_df.sort_index())
```

The two models have completely different definitions of what makes data "weird":

Isolation Forest (Machine Learning) looks globally: It flagged 12 total weeks. It looks at the entire 4-year dataset and says, "Any week over ~$20,000 is extremely rare compared to the overall average, so I am flagging it."

Z-Score (Statistical) looks locally: It only flagged 5 weeks. Because we used a 4-week rolling average, the Z-score only flags a week if it is a massive, sudden surprise compared to the weeks immediately before it.

```python
# Task 6: Product Demand Segmentation using Clustering
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import warnings
import os
warnings.filterwarnings('ignore')

print("--- Preparing Data for Clustering ---")
# Safely load fresh data
df = pd.read_csv('archive/train.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Year'] = df['Order Date'].dt.year

# 1. Feature Engineering per Sub-Category
# a. Total Sales & Average Order Value (AOV)
subcat_stats = df.groupby('Sub-Category').agg(
    Total_Sales=('Sales', 'sum'),
    AOV=('Sales', 'mean')
)

# b. Sales Volatility (Standard deviation of monthly sales)
monthly_subcat = df.set_index('Order Date').groupby('Sub-Category').resample('ME')['Sales'].sum().reset_index()
volatility = monthly_subcat.groupby('Sub-Category')['Sales'].std().rename('Volatility')

# c. YoY Growth (Comparing most recent full years: 2017 to 2018)
sales_17 = df[df['Year'] == 2017].groupby('Sub-Category')['Sales'].sum()
sales_18 = df[df['Year'] == 2018].groupby('Sub-Category')['Sales'].sum()
yoy_growth = ((sales_18 - sales_17) / sales_17).rename('YoY_Growth').fillna(0)

# Combine all features into our final clustering dataframe
cluster_df = subcat_stats.join(volatility).join(yoy_growth).fillna(0)

print("--- Scaling Data and Finding Optimal K (Elbow Method) ---")
# Standardize the features so large numbers (Total Sales) don't overpower small numbers (Growth Rate)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(cluster_df)

# Run K-Means for 1 to 10 clusters to find the "Elbow"
inertia = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method
plt.figure(figsize=(8, 4))
plt.plot(K_range, inertia, marker='o', color='purple', linewidth=2)
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig("charts/Elbow_Method_for_Optimal_K.png")
plt.show()

print("\n--- Applying K-Means (K=4) & PCA ---")
# Based on typical retail data, K=4 usually provides the most actionable business segments. 
# We apply K=4 here to generate the strategies required by the project brief.
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
cluster_df['Cluster'] = kmeans.fit_predict(scaled_data)

# Use PCA to reduce the 4 features into 2 dimensions for visualization
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
cluster_df['PCA1'] = pca_result[:, 0]
cluster_df['PCA2'] = pca_result[:, 1]

# Plot the clusters
plt.figure(figsize=(12, 7))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', palette='tab10', data=cluster_df, s=150, alpha=0.8)

# Annotate the points with Sub-Category names so the business can actually read it
for i, row in cluster_df.iterrows():
    plt.text(row['PCA1'] + 0.1, row['PCA2'] + 0.05, i, fontsize=10, weight='bold')

plt.title('Product Demand Segments (PCA 2D Projection)', fontsize=14)
plt.xlabel('PCA Component 1 (Primary Variance)')
plt.ylabel('PCA Component 2 (Secondary Variance)')
plt.legend(title='Demand Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

os.makedirs("charts", exist_ok=True)
plt.savefig("charts/Product_Segmentation_Clusters.png")
plt.show()

# Display Cluster Avg so we can figure out what each cluster actually represents
print("\n--- Cluster Profiles (Averages) ---")
cluster_profiles = cluster_df.groupby('Cluster')[['Total_Sales', 'AOV', 'Volatility', 'YoY_Growth']].mean()
display(cluster_profiles.round(2))
```
